```python
import pandas as pd
import matplotlib.pyplot as plt

CUSTOMERS = pd.read_csv("../01_Data/customers.csv")
INVOICES = pd.read_csv("../01_Data/invoices.csv")
SUBSCRIPTIONS = pd.read_csv("../01_Data/subscriptions.csv")
SUBSCRIPTION_PLANS = pd.read_csv("../01_Data/subscription_plans.csv")
CHURN_EVENTS = pd.read_csv("../01_Data/Churn_events.csv")

print("DATA LOADED")
```

```python
REVENUE_DATA = (
    INVOICES
    .merge(CUSTOMERS,on="customer_id",how="left")
    .merge(SUBSCRIPTIONS,on="customer_id",how="left")
    .merge(SUBSCRIPTION_PLANS,on="plan_id",how="left")
)

PLAN_REVENUE = (
    REVENUE_DATA
    .groupby("plan_name")["amount"]
    .sum()
    .sort_values(ascending=False)
)

PLAN_REVENUE.head()
```

```python
EXECUTIVE_KPIS = pd.DataFrame({

    "KPI":[
        "Total Revenue",
        "Revenue Customers",
        "Active Customers",
        "Churned Customers",
        "Churn Rate",
        "MRR",
        "ARR",
        "Revenue Per Active Customer",
        "NRR",
        "GRR"
    ],

    "VALUE":[
        3462189,
        8027,
        5469,
        2558,
        "32%",
        2617481,
        31409768,
        633.06,
        "58%",
        "53%"
    ]
})

EXECUTIVE_KPIS
```

```python
EXECUTIVE_KPIS.to_csv(
    "executive_kpi_summary.csv",
    index=False
)

print(
    "KPI SUMMARY EXPORTED"
)
```

```python
PLAN_REVENUE = (
    REVENUE_DATA
    .groupby("plan_name")["amount"]
    .sum()
    .sort_values(ascending=False)
)

print("REVENUE BY PLAN")
print(PLAN_REVENUE)

PLAN_REVENUE.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Revenue Drivers by Plan")
plt.xlabel("Subscription Plan")
plt.ylabel("Revenue")

plt.show()
```

```python
CHURN_EVENTS[
    "churn_reason"
].value_counts().plot(
    kind="bar"
)

plt.title(
    "Top Churn Drivers"
)

plt.show()
```
